```python
def convert_units():
    try:
        print("Choose conversion type:")
        print("1. Length")
        print("2. Weight")
        print("3. Volume")

        choice = input("Enter your choice (1/2/3): ")

        value = float(input("Enter the value to convert: "))

        # LENGTH CONVERSION
        if choice == "1":
            print("\nLength Conversion")
            print("1. Meters to Feet")
            print("2. Feet to Meters")

            option = input("Choose option (1/2): ")

            if option == "1":
                result = value * 3.28084
                print(value, "meters =", result, "feet")

            elif option == "2":
                result = value / 3.28084
                print(value, "feet =", result, "meters")

            else:
                print("Invalid option.")

        # WEIGHT CONVERSION
        elif choice == "2":
            print("\nWeight Conversion")
            print("1. Kilograms to Pounds")
            print("2. Pounds to Kilograms")

            option = input("Choose option (1/2): ")

            if option == "1":
                result = value * 2.20462
                print(value, "kg =", result, "pounds")

            elif option == "2":
                result = value / 2.20462
                print(value, "pounds =", result, "kg")

            else:
                print("Invalid option.")

        # VOLUME CONVERSION
        elif choice == "3":
            print("\nVolume Conversion")
            print("1. Liters to Gallons")
            print("2. Gallons to Liters")

            option = input("Choose option (1/2): ")

            if option == "1":
                result = value * 0.264172
                print(value, "liters =", result, "gallons")

            elif option == "2":
                result = value / 0.264172
                print(value, "gallons =", result, "liters")

            else:
                print("Invalid option.")

        else:
            print("Invalid conversion type.")

    except ValueError:
        print("Error: Please enter a numeric value.")
```

```python
def find_sum(numbers):
    return sum(numbers)

def find_average(numbers):
    return sum(numbers) / len(numbers)

def find_maximum(numbers):
    return max(numbers)

def find_minimum(numbers):
    return min(numbers)

try:
    user_input = input("Enter numbers separated by spaces: ")
    numbers = [float(x) for x in user_input.split()]

    print("Choose operation:")
    print("1 Sum")
    print("2 Average")
    print("3 Maximum")
    print("4 Minimum")

    choice = input("Enter choice: ")

    if choice == "1":
        print("Sum =", find_sum(numbers))

    elif choice == "2":
        print("Average =", find_average(numbers))

    elif choice == "3":
        print("Maximum =", find_maximum(numbers))

    elif choice == "4":
        print("Minimum =", find_minimum(numbers))

    else:
        print("Invalid choice")

except:
    print("Invalid input")
```

```python
def flatten(lst):
    result = []
    for sub in lst:
        for item in sub:
            result.append(item)
    return result

print(flatten([[1,2],[3,4],[5]]))
```

```python
def sum_nested(lst):
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total

print(sum_nested([[1,2],[3,[4,5]],6]))
```

```python
def remove_element(lst, elem):
    result = []
    for sub in lst:
        new = []
        for x in sub:
            if x != elem:
                new.append(x)
        result.append(new)
    return result

print(remove_element([[1,2],[3,2],[4,5]],2))
```

```python
def find_max(lst):
    max_value = float('-inf')

    for item in lst:
        if isinstance(item, list):
            value = find_max(item)
            if value > max_value:
                max_value = value
        else:
            if item > max_value:
                max_value = item

    return max_value


print(find_max([[1,2],[3,[4,5]],6]))
```

```python
def count_occurrences(lst, elem):
    count = 0
    for sub in lst:
        count += sub.count(elem)
    return count

print(count_occurrences([[1,2],[2,3],[2,4]],2))
```

```python
def deep_flatten(lst):
    result = []
    for item in lst:
        if isinstance(item,list):
            result += deep_flatten(item)
        else:
            result.append(item)
    return result

print(deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
```

```python
def average_nested(lst):
    flat = deep_flatten(lst)
    return sum(flat)/len(flat)

print(average_nested([[1,2],[3,4],[5,6]]))
```
